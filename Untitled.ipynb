```python
import os
import matplotlib.pyplot as plt
```

```python
path = './progress/train_v2/imgs/log.txt'
```

```python
a = {}
lr = []
tr_loss = []
val_loss = []
val_auc = []
val_acc = []
val_sen = []
val_spe = []
with open(path, 'r') as f:
    for line in f:
        #print(line)
        for idx, i in enumerate(line.split(',')):
            a, b = i.split(':')
        
            #print(a.replace('{', ''), b.replace('}', ''))
            if idx==0:
                lr.append(float(b.replace('}', '')))
            elif idx ==1:
                tr_loss.append(float(b.replace('}', '')))
            elif idx==3:
                val_loss.append(float(b.replace('}', '')))
            elif idx==4:
                val_auc.append(float(b.replace('}', '')))
            elif idx==5:
                val_acc.append(float(b.replace('}', '')))
            elif idx==6:
                val_sen.append(float(b.replace('}', '')))
            elif idx==7:
                val_spe.append(float(b.replace('}', '')))
            
        
```
